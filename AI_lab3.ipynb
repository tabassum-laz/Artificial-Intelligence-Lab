```python
x = df["weight"].mean()
print("The mean is ",x)
new_df = df.fillna(x)
print(new_df)
print(x)
```

```python
x = df["weight"].median()
print("The median is ",x)
new_df = df.fillna(x)
print(new_df)
print(x)
```

```python
x = df["weight"].mode() [0]
print("The mode is ",x)
new_df = df.fillna(x)
print(new_df)
print(x)
```
