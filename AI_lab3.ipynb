```python
new_df = df.dropna()
print(df)
print(new_df)
```

```python
new_df = df.fillna(50)
print(new_df)
```

```python
x = df["weight"].mean()
print("The mean is ",x)
new_df = df.fillna(x)
print(new_df)
print(x)
```

```python
x = df["weight"].median()
print("The median is ",x)
new_df = df.fillna(x)
print(new_df)
print(x)
```

```python
x = df["weight"].mode() [0]
print("The mode is ",x)
new_df = df.fillna(x)
print(new_df)
print(x)
```
